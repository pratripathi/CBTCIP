```python
df=pd.read_csv('book4.csv')
```

```python
df["Gender"] = df["Gender"].map({0:"Male", 1:"Female"})
df["GeneticRisk"] = df["GeneticRisk"].map({0:"Low", 1:"Medium", 2:"High"})
df["Hypertension"] = df["Hypertension"].map({0:"No", 1:"Yes"})
```

```python
sns.countplot(x='Gender',data=df,hue='GeneticRisk')
plt.title('countplot of gender')
```

```python
sns.countplot(x='Smoking', data=df, hue = "Hypertension")
plt.title('Countplot of smoker')
plt.show()
```

```python
sns.pairplot(df[['Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'LiverFunctionTest', 'Diagnosis']], hue='Diagnosis')
plt.show()
```

```python
df["Age"].dtype, df["BMI"].dtype
```

```python
sns.scatterplot(x= df["Age"], y = df["PhysicalActivity"], hue=df["Smoking"], style=df["GeneticRisk"])
plt.xlabel("Age")
plt.ylabel("Activity")
plt.title("Age vs Physicical activity")
plt.legend()
plt.show()

```
